```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the file path
file_path = '/content/drive/MyDrive/Datasets/Athlete Data.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Sport'], df['Sex'])

# Display the contingency table
print("Contingency Table:")
print(contingency_table)

# Perform the Chi-square test for independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("\nChi-square test statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("\nExpected frequencies table:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# Interpret the results
alpha = 0.05
print("\nSignificance level:", alpha)
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between Sport and Sex.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between Sport and Sex.")
```
